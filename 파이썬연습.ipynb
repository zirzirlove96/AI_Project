```python
#!pip install pandas
#!pip install --upgrade numpy
import pandas as pd
```

```python
document = ["코로나 상생지원금 문의입니다."
, "낮12시~2시 지하철 운행시간 문의입니다."
,"밤 12시 이후 버스 운행시간 문의입니다."
,"사회적 거리두기로 인한 10시 이후 영업시간 안내입니다."
,"Bus 운행시간 문의입니다."
,"Taxi 운행시간 문의입니다."]

df = pd.DataFrame(document, columns=["문서"])
#pd.DataFrame({"문서" : document})
#pd.Series(document)
```

```python
df.shape
#데이터프레임의 정보
df.info()
#문의날짜
df["날짜"] = "2022-05-01" #날짜라는 컬럼이 생성되며 날짜 컬럼의 데이터는 2022-05-01로 채워진다.
df
```

```python
#df["문서"] #Series형태
df[["문서","날짜"]]#series 형식으로 문서,날짜 모두 가져오고 싶을때 list 형식으로01
```

```python
#특정 행 가져오기  **(loc)
#df.loc[0]
#type(df.loc[0])
#df.loc[[0,1]] # 0~1번째 행 가져오기

#(1,0)에 대한 데이터 가져오기
#df.loc[1,"문서"]

#문서 열에 대한 0~2번째 행의 데이터가져오기
df.loc[[0,2],"문서"]

#날짜 열에 대한 2~3번째 행의 데이터가져오기
df.loc[[2,3], "날짜"]

#0~2번째 행의 전체 데이터 가져오기
df.loc[[0,2]]
df.loc[[0.2],["문서","날짜"]]
```

```python
#.iloc를 사용해서 일부 슬라이싱 하기
#(1,1)에 데이터 가져오기
#인덱스 순서대로 가져올 수 있다.
df.iloc[1,1]

#loc와 같이 dataframe형식으로 가져올 수 있음
#df.iloc[[0,1],[0,1]]

#2개의 행까지만, 1개의 열까지만(슬라이싱)
#df.iloc[:2,:1]

#loc로도 가져올 수 있음
df.loc[0,"문서"]
```
